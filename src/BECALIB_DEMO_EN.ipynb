## Import classes an methods from BECALIB library

```python
from becalib import MaterialLayer, AirLayer
from becalib import Component
```

## Layers setup
Instantiate a layer objet for each component layer \
Example below

```python
concrete  =  MaterialLayer(
            name="Concrete",
            thickness=0.15, # m
            thermal_conductivity=1.8, # lambda W/mK
            specific_heat_capacity=1000, #  c J/kgK
            gross_density=2400, # ro kg/mc
        )

air_gap  =  AirLayer(
    name="Air gap",
    thickness=0.1, # m
    heat_flow_direction="Do"

)

bricks  =  MaterialLayer(
    name="Bricks",
    thickness=0.12, # m
    thermal_conductivity=0.8, # lambda W/mK
    specific_heat_capacity=840, #  c J/kgK
    gross_density=1800, # ro kg/mc

)
insulation  =  MaterialLayer(
    name="Insulation",
    thickness=0.05, # m
    thermal_conductivity=0.035, # lambda W/mK
    specific_heat_capacity=840, #  c J/kgK
    gross_density=175, # ro kg/mc

)

plaster  =  MaterialLayer(
    name="Plaster",
    thickness=0.02, # m
    thermal_conductivity=0.9, # lambda W/mK
    specific_heat_capacity=840, #  c J/kgK
    gross_density=1400, # ro kg/mc

)
```

## Components setup:
Instantiate a component objet and set:
* List of layers  (interior to exterior)
* Heat flow direction:
        "Ho": Horizontal (example: wall) or
        "Up": Upwards (example Roof) or
        "Do": Downwards (example floor)
* language (en, fr, etc..)

```python
wall = Component(name="wall",
                layers=[
                concrete,
                air_gap,
                bricks,
                insulation,
                plaster ],
                heat_flow_direction="Ho",
                language="en"
                )
```

## Get a component table (Pandas DataFrame)

```python
wall.get_layers_dataframe(data_type="st")
```

## Print values

```python
print(wall.get_summer_performance_key_values())
```
